```python
import numpy as np
import pandas as pd
data={
    'ID':[1,2,3,4,5],
    'Name':['Ram', 'Hari','Mohan','Shyam','John'],
    'Age':[22,44,33,np.nan,23],
    'Location':['Kathmandu', 'pokhara', 'kathmandu','Butwal','chitwan'],
    'Salary':[50000,20000,45000,19000,30000]
}
df=pd.DataFrame(data)
df.head()
```

```python
x=df['Age'].mean()
print(x)
```

```python
df['Age'].fillna(x, inplace=True)
print(df)
```

```python
y=df['Salary'].median()
print(y)
```

```python
df['Bonus']=df['Salary']*0.15
df
```

```python
df['Location']=df['Location'].str.upper()
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.bar(df['Name'],df['Salary'])
plt.show()
```

```python
new=df.groupby('Location')['Salary'].mean().reset_index()
print(new)
```
